```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('insurance_claims.csv')
df
```

```python
df.replace('?', np.nan, inplace = True)
```

```python
df
```

```python
df['collision_type'] = df['collision_type'].fillna(df['collision_type'].mode()[0])
df['property_damage'] = df['property_damage'].fillna(df['property_damage'].mode()[0])
df['police_report_available'] = df['police_report_available'].fillna(df['police_report_available'].mode()[0])
df
```

```python
to_drop = ['policy_number','policy_bind_date','policy_state','insured_zip','incident_location','incident_date',
           'incident_state','incident_city','insured_hobbies','auto_make','auto_model','auto_year', '_c39']

df.drop(to_drop, inplace = True, axis = 1)
df
```

```python
plt.figure(figsize = (18, 12))

corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype = bool))

sns.heatmap(data = corr, mask = mask, annot = True, fmt = '.2g', linewidth = 1)
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
categorical = x.select_dtypes(include = ['object'])
categorical
```

```python
categorical = pd.get_dummies(categorical, drop_first = True)
categorical
```

```python
education_columns = [
    'insured_education_level_College',
    'insured_education_level_High School',
    'insured_education_level_JD',
    'insured_education_level_MD',
    'insured_education_level_Masters',
    'insured_education_level_PhD'
]
education_counts = categorical[education_columns].apply(lambda col: col.value_counts())
for column in education_columns:
    print(f"Counts for {column}:\n{education_counts[column].to_string(index=False)}\n")
```

```python
num_df = x.select_dtypes(include = ['int64'])
x = pd.concat([num_df, categorical], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
num_df1 = x_train.select_dtypes(include = ['int64'])
x = pd.concat([num_df, categorical], axis = 1)
num_df1
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(num_df1)
```

```python
scaled_data
```

```python
scaled_num_df = pd.DataFrame(data = scaled_data, columns = num_df.columns, index = x_train.index)
scaled_num_df
```

```python
x_train.drop(columns = scaled_num_df.columns, inplace = True)
x_train = pd.concat([scaled_num_df, x_train], axis = 1)
x_train
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)*100
print(f"accuracy score is: {round(acc, 2)}%")
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
y_pred1 = dtc.predict(x_test)
```

```python
acc1 = accuracy_score(y_test, y_pred1)*100
print(f"accuracy score is: {round(acc1, 2)}%")
```

```python
from sklearn.model_selection import GridSearchCV
grid_params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [3, 5, 7, 10],
    'min_samples_split' : range(2, 10, 1),
    'min_samples_leaf' : range(2, 10, 1)
}
```

```python
grid_search = GridSearchCV(dtc, grid_params, cv = 8, n_jobs = -1, verbose = 1)
grid_search.fit(x_train, y_train)
```

```python
dtc = grid_search.best_estimator_
y_pred2 = dtc.predict(x_test)
```

```python
acc2 = accuracy_score(y_test, y_pred2) *100
print(f"The accuracy score is {round(acc2, 2)}%")
```
